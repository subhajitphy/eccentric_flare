```python
from gwecc import waveform
from constants import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
tarr=np.linspace(0,10*yr,1000)
period = 1.5*yr;inc = 0;M0=1e9;
q=1;
eta=q/(1+q)**2;
n0 = 2*np.pi/period

e0 = 0.8;distGW=1
order=3
```

```python
hp,hx=waveform(tarr,n0,M0,e0,eta,inc,distGW,order)
```

```python
plt.figure(figsize=(10,5))

plt.plot(tarr/yr,hp,label='hp')
plt.plot(tarr/yr,hx,label='hx')
plt.xlim(min(tarr/yr),max(tarr/yr))
plt.legend()
plt.grid()
plt.show()
```

```python
from gwecc import add_ecc_cgw
```

```python
def hms_to_rad(hh, mm, ss):
    sgn = np.sign(hh)
    return sgn * (sgn * hh + mm / 60 + ss / 3600) * np.pi / 12


def dms_to_rad(dd, mm, ss):
    sgn = np.sign(dd)
    return sgn * (sgn * dd + mm / 60 + ss / 3600) * np.pi / 180
```

```python
RA_GW = hms_to_rad(8, 0, 0)
DEC_GW = dms_to_rad(-20, 0, 0)
```

```python
psrra=hms_to_rad(4, 37, 0)
psrdec=dms_to_rad(-47, 15, 0)
```

```python
pdist1=156.79*pc
pdist2=157.04*pc
#Mc=1e9*eta**(3/5)

pdist=1e9*pc
```

```python
res1=add_ecc_cgw(toas=tarr,
    theta=np.pi/2-psrdec,
    phi=psrra,
    cos_gwtheta=np.sin(DEC_GW),
    gwphi=RA_GW,
    psi=0,
    cos_inc=np.cos(inc),
    log10_n=np.log10(n0),
    q=q,
    e0=e0,
    log10_M=np.log10(M0),
    tref=0,
    pdist=pdist1,
    distGW=distGW,res='Both')

res2=add_ecc_cgw(toas=tarr,
    theta=np.pi/2-psrdec,
    phi=psrra,
    cos_gwtheta=np.sin(DEC_GW),
    gwphi=RA_GW,
    psi=0,
    cos_inc=np.cos(inc),
    log10_n=np.log10(n0),
    q=q,
    e0=e0,
    log10_M=np.log10(M0),
    tref=0,
    pdist=pdist2,
    distGW=distGW,res='Both')
```

```python
plt.figure(figsize=(10,5))
plt.plot(tarr/yr,res1)
plt.plot(tarr/yr,res2)
```
